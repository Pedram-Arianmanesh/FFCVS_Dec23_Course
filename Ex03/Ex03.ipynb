## Find max of Total.Population and subtract all values of Total.Population column from the maximum value using lambda function.

# Exercise 15
## Implement following architecture using OOP!
### use __str__ to print output same as: Breed of the dog is {Breed}, It is {Size} and {Color} , the dog is {Age} years old.



```python
class Dog():
    def __init__(self, breed, size, color, age):
        self.breed = breed
        self.size = size
        self.color = color
        self.age = age
    def __str__(self):
        return f'Breed of the dog is {self.breed}, It is {self.size} and {self.color} , the dog is {self.age} years old.'

Bella = Dog('Beaglier', 'small', 'red', 2)
Max = Dog('Foxhound', 'big', 'white', 3)
Cooper = Dog('Beagle', 'medium', 'brown', 2)

print('Bella\'s info:', Bella)
print('Max\'s info:', Max)
print('Cooper\'s info:', Cooper)
```
