## Implement following architecture using OOP!
### use __str__ to print output same as: Desired text, using input data.


```python
# declare parent class
class Animal:
    def __init__(self, category, breed, size, age, color):
        self.category = category
        self.breed = breed
        self.size = size
        self.age = age
        self.color = color
        
    def eat(self):
        return f'{self.category} animal can eat.'
    
    def sleep(self):
        return f'{self.category} animal can sleep.'
    
    def sit(self):
            return f'{self.category} animal can sit.'
    
    def run(self):
        return f'{self.category} animal can run.'

    def general(self):
        return f'Category of this animal is {self.category} and its breed is {self.breed}, It is {self.size} and {self.color}, the animal is {self.age} years old.'

```

```python
# declare child classes

class Pet(Animal):
    def __init__(self, category, breed, size, age, color, name, owner, HC_num):
        self.category = category
        self.breed = breed
        self.size = size
        self.age = age
        self.color = color
        self.name = name
        self.owner = owner
        self.HC_num = HC_num
        
    def hug(self):
        return f'You can hug {self.name}.'
    
    def bring(self):
        return f'You can bring {self.name} with your self.'

    def __str__(self):
        return f'{self.name} is name of this animal. Its health card number is: {self.HC_num} and its owner is {self.owner}.'

class Wild(Animal):
    def __init__(self,category, breed, size, age, color, jungle_name):
        self.category = category
        self.breed = breed
        self.size = size
        self.age = age
        self.color = color
        self.jungle_name = jungle_name
    
    def kill(self):
            return f'{self.category} animal can kill!'
        
    def hunt(self):
        return f'{self.category} animal can hunt.'
    
    def __str__(self):
        return f'This {self.breed} lives in {self.jungle_name} jungle.'

    
```

```python
pet_example = Pet(category='Felidae', breed='Persian cat', size='small', age='2', color='white', name='Tom', owner='Asqar', HC_num='25867')
print(pet_example.general())
print(str(pet_example))
```

```python
wild_example = Wild(category='Felidae', breed='Lion', size='big', age='7', color='brown', jungle_name='Zoria')
print(wild_example.general())
print(str(wild_example))
```
